# 퀘스트 1번: 은행계좌 만들기


## Q1. Account 클래스 & 생성자 / Q2. 클래스 변수


```python
import random

class Account:
    """은행 계좌 클래스"""

    # Q2. 클래스 변수 - 생성된 계좌 객체의 개수 저장
    account_count = 0

    # Q1. 생성자 (__init__ 매직메서드)
    def __init__(self, owner, balance):
        self.bank_name = "SC은행"                      # 은행이름 고정
        self.owner = owner                              # 예금주
        self.balance = balance                          # 초기 잔액
        self.deposit_count = 0                          # Q7. 입금 횟수
        self._deposit_log = []                          # Q10. 입금 내역
        self._withdraw_log = []                         # Q10. 출금 내역

        # 계좌번호 랜덤 생성 (3자리-2자리-6자리)
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        self.account_number = f"{part1}-{part2}-{part3}"

        Account.account_count += 1                      # Q2. 계좌 수 증가

    # Q3. 클래스 변수 출력 메서드
    def get_account_num(self):
        """생성된 계좌 객체의 개수를 출력"""
        print(f"개설된 계좌 수: {Account.account_count}")

    # Q4. 입금 메서드
    def deposit(self, amount):
        """입금 - 최소 1원 이상만 가능"""
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.balance += amount
        self.deposit_count += 1
        self._deposit_log.append(amount)
        print(f"{amount:,}원이 입금되었습니다.")

        # Q7. 입금 횟수가 5회가 될 때마다 1% 이자 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"★ 입금 {self.deposit_count}회 달성! 이자 {interest:,}원이 지급되었습니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        """출금 - 잔고 이상 출금 불가"""
        if amount > self.balance:
            print(f"잔고가 부족합니다. (현재 잔고: {self.balance:,}원)")
            return
        self.balance -= amount
        self._withdraw_log.append(amount)
        print(f"{amount:,}원이 출금되었습니다.")

    # Q6. 정보 출력 메서드
    def display_info(self):
        """계좌 정보 출력 (잔고는 세 자리마다 쉼표)"""
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, "
              f"계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    # Q10. 입금 내역 출력
    def deposit_history(self):
        """입금 내역을 출력"""
        print(f"[ {self.owner} 입금 내역 ]")
        if not self._deposit_log:
            print("  입금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self._deposit_log, 1):
                print(f"  {i}회차: {amount:,}원")

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        """출금 내역을 출력"""
        print(f"[ {self.owner} 출금 내역 ]")
        if not self._withdraw_log:
            print("  출금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self._withdraw_log, 1):
                print(f"  {i}회차: {amount:,}원")

print("✅ Account 클래스 정의 완료!")
```

## Q3. 클래스 변수 출력 테스트


```python
# 계좌 생성 테스트
acc1 = Account("파이썬", 1_000_000)
acc1.display_info()

acc1.get_account_num()
```

## Q4. 입금 메서드 테스트


```python
# 정상 입금
acc1.deposit(50_000)

# 비정상 입금 (1원 미만)
acc1.deposit(0)

acc1.display_info()
```

## Q5. 출금 메서드 테스트


```python
# 정상 출금
acc1.withdraw(200_000)

# 비정상 출금 (잔고 초과)
acc1.withdraw(99_999_999)

acc1.display_info()
```

## Q6. 정보 출력 메서드 테스트


```python
acc1.display_info()
```

## Q7. 이자 지급 테스트


```python
# 현재 입금 횟수 확인 (이미 1회 입금됨)
print(f"현재 입금 횟수: {acc1.deposit_count}회")
print()

# 5회가 될 때까지 추가 입금
for _ in range(4):
    acc1.deposit(10_000)

print()
acc1.display_info()
```

## Q8. 여러 객체 생성


```python
# 기존 카운트 리셋 (테스트 편의)
Account.account_count = 0

accounts = [
    Account("파이썬", 1_000_000),
    Account("자바", 500_000),
    Account("코틀린", 2_000_000),
    Account("자바스크립트", 800_000),
]

print("=" * 55)
for acc in accounts:
    acc.display_info()
print("=" * 55)
accounts[0].get_account_num()
```

## Q9. 객체 순회 — 잔고 100만 원 이상 고객만 출력


```python
print("[ 잔고 100만 원 이상 고객 ]\n")

for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()
```

## Q10. 입금/출금 내역 기록 및 출력

```python
# 거래 수행
accounts[0].deposit(100_000)
accounts[0].deposit(200_000)
accounts[0].withdraw(50_000)
accounts[0].withdraw(30_000)

print()
accounts[0].deposit_history()
print()
accounts[0].withdraw_history()
```
